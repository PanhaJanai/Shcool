# Pandas

## Series

```python
import pandas as pd

s = pd.Series([1, 2, 3, 4, 5], index=["Apple", "Banana", "Coconut", "Dragonfruit", "Elderberry"])
print(s)

print()

marvel_year = {
    "Iron Man": 2010,
    "Captain America": 2011,
    "Thor": 2013,
    "Winter Soldier": 2014,
    "Ultron": 2015
}

s = pd.Series(marvel_year)

print(s)
print()
print(s.index)
```

## pandas X numpy

```python
import pandas as pd
import numpy as np

np_arr = np.arange(0, 5)
s = pd.Series(np_arr)

print(s)
print()

date_arr = pd.date_range("2023-12-27", "2023-12-31")

print(date_arr)
print()

s = pd.Series([x for x in marvel_year], index=date_arr)

print(s)
```
