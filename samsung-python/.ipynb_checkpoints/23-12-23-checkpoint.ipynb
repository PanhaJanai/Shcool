# DataFrame

```python
import pandas as pd

# Creating DataFrame
data = {
    "A": [1, 2, 3, 4],
    "B": ['a', 'b', 'c', 'd']
}
print(pd.DataFrame(data))

# Creating DataFrame with pandas' series
s1 = pd.Series([1.0, 2.0, 3.0], index=['a', 'b', 'c'])
s2 = pd.Series([1.0, 2.0, 3.0, 4.0], index=['a', 'b', 'c', 'd'])

data2 = {"column one": s1, "column two": s2}
print(pd.DataFrame(data2))

# Changing DataFrame's index and columns name
df = pd.DataFrame(data=data2)
df.index = ["index 0", "index 1", "index 2", "index 3"]
print(df)

df.columns = ["column 1", "column 2"]
print(df)
```

```python
import pandas as pd

# Rename index and column name
data = {
    "A": [1, 2, 3, 4],
    "B": ['a', 'b', 'c', 'd']
}
df2 = pd.DataFrame(data)

df2 = df2.rename(index={0: "Index 0"})
df2 = df2.rename(columns={'A': "Numbers"})

print(df2)
```
