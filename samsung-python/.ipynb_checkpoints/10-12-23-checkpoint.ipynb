```python
def find_min(row, col, triangle):
    if row == len(triangle):
        return 0
    else:
        mini = min(find_min(row + 1, col, triangle),
                  find_min(row + 1, col + 1, triangle))
        return triangle[row][col] + mini

triangle = [
    [2],
    [3, 4],
    [6, 5, 7],
    [4, 1, 8, 3]
]

mini = find_min(0, 0, triangle)
print("The minimum cost is", mini)
```

## Binary Search with iteration

```python
def bin_search_iter(nums, x):
    low, high = 0, len(nums) - 1
    while low <= high:
        mid = (low + high) // 2
        if nums[mid] == x:
            return mid
        elif nums[mid] > x:
            high = mid - 1
        else:
            low = mid + 1
    return -1

l = [5, 6, 1, 2, 3, 7, 8]
index = bin_search_iter(l, 7)
print(l[index])
```

## Binary Search with Divide-and-conquer

```python
def bin_search_div_n_con(nums, x, low=0, high=0):
    if low > high:
        return -1
    else: 
        mid = (low + high) // 2
        if nums[mid] == x:
            return mid
        elif nums[mid] > x:
            return bin_search_div_n_con(nums, x, low, mid - 1)
        else:
            return bin_search_div_n_con(nums, x, mid + 1, high)
        
l = [5, 6, 1, 2, 3, 7, 8]
index = bin_search_div_n_con(l, 2, low=2, high=4)
print(l[index])
```

## Fibonacci without memoization

## Fibonacci with memoization

```python
c1, c2 = 0, 0

f = {0: 0, 1: 1}

def fib_memoi(n):
    global c1
    c1 += 1
    if n in f:
        return f[n]
    else:
        f[n] = fib_memoi(n - 1) + fib_memoi(n - 2)
        return f[n]
    
def fibMemoi2(f, n):
    global c2
    c2 += 1
    if n < 2:
        return f[n]
    else:
        if f[n] < 0:
            f[n] = fibMemoi2(f, n - 1) + fibMemoi2(f, n - 2)
        return f[n]
    
user = int(input("Please enter a number: "))
print(fib_memoi(user))

f = [0, 1] + [-1] * (user - 1)
print(f)
print(fibMemoi2(f, user))
print(f)

print(c1, c2)
```

## Pascal triangle with recursion 

```python
def binomialRecursion(n, k):
    if k == 0 or n == k:
        return 1
    else:
        return binomialRecursion(n - 1, k - 1) + binomialRecursion(n - 1, k)
    
n = int(input("Input the value of n: "))
k = int(input("Input the value of k: "))
print(f"Binomial({n}, {k} is {binomialRecursion(n, k)})")
```

## Pascal's triangle by dynamic programming

```python
def binomialDynamic(n, k):
    b = [[0] * (i + 1) for i in range(n + 1)]
    for i in range(n + 1):
        for j in range(i + 1):
            if j == 0 or j == i:
                b[i][j] = 1
            else:
                b[i][j] = b[i - 1][j - 1] + b[i - 1][j]
    return b[n][k]

n = int(input("Input the value of n: "))
k = int(input("Input the value of k: "))
print(f"Binomial({n}, {k} is {binomialDynamic(n, k)})")
```
